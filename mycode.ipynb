```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer

# Read text file
with open("Sherlock Holmes.txt", "r", encoding="utf-8") as f:
    data = f.read()

# Tokenize
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])

print("Vocabulary:", (tokenizer.word_index))
```

```python
print(len(tokenizer.word_index))
```

```python
import re
sentences = re.split(r'[.\n]', data)
# print(sentences)
sentences = [s.strip() for s in sentences if s.strip()]
print(sentences)
```

```python
input_seq=[]
for s in sentences:
    tokenized_sen=tokenizer.texts_to_sequences([s])[0]
    # print(tokenized_sen)
    for i in range(1,len(tokenized_sen)):
        input_seq.append(tokenized_sen[:i+1])
```

```python
input_seq
```

Padding

```python
max_length=max([len(x) for x in input_seq])
```

```python
print(max_length)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_seq=pad_sequences(input_seq,maxlen=max_length,padding='pre')
```

```python
padded_input_seq
```

```python
x=padded_input_seq[:,:-1]
x
```

```python
y=padded_input_seq[:,-1]
y
```

```python
x.shape
```

```python
y.shape
```

```python
vocab_size = len(tokenizer.word_index) + 1
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=100, input_length=x.shape[1]),
    LSTM(128),
    Dense(vocab_size, activation='softmax')
])


model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

```

```python
model.summary()
```

```python
model.fit(x,y,epochs=20)
```

```python
import time
import numpy as np
text="what is your"
for i in range(10):
    token_text=tokenizer.texts_to_sequences([text])[0]
    padded_token_text=pad_sequences([token_text],maxlen=16,padding='pre')
    pos=np.argmax(model.predict(padded_token_text))
    for word,index in tokenizer.word_index.items():
        if index==pos:
            text=text+" "+word
            print(text)
            time.sleep(2)
```
